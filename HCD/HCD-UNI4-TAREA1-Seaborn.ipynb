Realiza un mapa de calor mediante la correlación del dataframe.

```python
# Seaborn
# Exercise 1

import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns

# Read Apple quote file
AAPL = pd.read_csv("AAPL.csv")
print(AAPL.head(5))
```

```python
# Correlación
sns.heatmap(AAPL.corr(), cmap="YlGnBu", annot = True)
```
