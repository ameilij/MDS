## Ejercicio Bokeh 1

Realiza una gráfica de toda la serie temporal sobre los precios más altos y bajos. Para que se diferencien más las líneas, resta 40 puntos a todos los precios más bajos.

```python
# Bokeh
# Exercise 21

import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
import numpy as np

output_notebook()

# Read Apple quote file
AAPL = pd.read_csv("AAPL.csv")
print(AAPL.head(5))
```

```python
p = figure(plot_width = 500, plot_height = 500)
days = np.arange(1, len(AAPL)+1)

p.circle(days, AAPL['Low']-40,line_color = 'red')
p.square(days, AAPL['High'], line_color = 'blue')
show(p)
```
