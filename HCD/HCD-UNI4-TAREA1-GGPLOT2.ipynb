## Ejercicio GGPLOT2

Realiza una nube de puntos con el volumen de acción y el cierre ajustado. Muestra, además, la línea de ajuste (regresión lineal), entre ambas variables. Utiliza un gradiente de color mediante el precio de cierre ajustado.

```python
# GGPLOT2
# Exercise 2

import pandas as pd
from plotnine import *


# Read Apple quote file
AAPL = pd.read_csv("AAPL.csv")
print(AAPL.head(5))

# Build a scatter plot
ggplot(AAPL) + geom_point(aes(x='Adj Close', y = 'Volume', fill = 'Adj Close')) + labs(title = "Cierre y Volumen Acciones APPL", x = "Cierre Ajustado", y = "Volumen") + geom_smooth(aes(x='Adj Close', y = 'Volume')) 
```
