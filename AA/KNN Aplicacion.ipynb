# Aplicación de KNN

```python
# librerias
from sklearn.datasets import make_blobs
import numpy as np
import mglearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# generar dataset
X, y = make_blobs(centers=2, random_state=5, n_samples=200)
plt.scatter(X[y == 0,0], X[y == 0,1], s=40, label="Clase Y=0", marker="^")
plt.scatter(X[y == 1,0], X[y == 1,1], s=40, label="Clase Y=1", marker="*", color="maroon")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

Se obtiene la frontera de separación de un KNN utilizando el valor 1 como número de vecinos, es decir, previsiblemente sobreajustando.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, stratify=y, random_state=5)

clf = KNeighborsClassifier(n_neighbors=1)
clf.fit(X_train, y_train)
```

```python
plt.scatter(X[y == 0,0], X[y == 0,1], s=40, label="Clase Y=0", marker="^")
plt.scatter(X[y == 1,0], X[y == 1,1], s=40, label="Clase Y=1", marker="*", color="orange")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
mglearn.plots.plot_2d_separator(clf, X)
plt.show()
```

```python
print('- Accuracy media en el conjunto de train: {}'.format(clf.score(X_train, y_train)))
print('- Accuracy media en el conjunto de test: {}'.format(clf.score(X_test, y_test)))
```

Por último, se prueba con el valor 15 como número de vecinos. Como se ve en la figura, en este caso la frontera se separa mejor y la métrica en test es mucho mejor, al no sobreajustar. 

```python
clf = KNeighborsClassifier(n_neighbors=15)
clf.fit(X_train, y_train)

plt.scatter(X[y == 0,0], X[y == 0,1], s=40, label="Clase Y=0", marker="^")
plt.scatter(X[y == 1,0], X[y == 1,1], s=40, label="Clase Y=1", marker="*", color="orange")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
mglearn.plots.plot_2d_separator(clf, X)
plt.show()
```

El algoritmo KNN se puede utilizar cuando el dataset no es muy grande, ya que es un algoritmo muy costoso computacionalmente. Además, suele obtener buenos resultados en casos con una dimensionalidad no muy alta; en caso contrario, el cálculo de vecinos según sus distancias puede no ser tan claro.
