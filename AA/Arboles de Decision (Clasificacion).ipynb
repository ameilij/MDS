# Árboles de decisión
En el caso de los árboles de decisión aplicados a problemas de clasificación, el funcionamiento es el mismo que en regresión. El único cambio es que, a la hora de hacer los cortes o decisiones, se toma como criterio de corte óptimo un concepto denominado ganancia de información.

```python
# librerias
from sklearn.datasets import make_blobs
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV, train_test_split
import mglearn
```

```python
# generar dataset
X, y = make_blobs(centers=2, random_state=434, n_samples=1000)
X, y = make_blobs(centers=2, random_state=5, n_samples=200)

plt.scatter(X[y == 0,0], X[y == 0,1], s=40, label="Clase Y=0", marker="^")
plt.scatter(X[y == 1,0], X[y == 1,1], s=40, label="Clase Y=1", marker="x", color="pink")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

```python
# entrenamiento
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=5)
depth = 3
clf = DecisionTreeClassifier(max_depth=depth)
clf.fit(X_train, y_train)
y_train_pred = clf.predict(X_train)
y_test_pred = clf.predict(X_test)
```

```python
plt.scatter(X[y == 0,0], X[y == 0,1], s=40, label="Clase Y=0", marker="+")
plt.scatter(X[y == 1,0], X[y == 1,1], s=40, label="Clase Y=1", marker="o", color="red")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
mglearn.plots.plot_2d_separator(clf, X)
plt.show()
```

```python
print('- Accuracy media en el conjunto train: {}'.format(clf.score(X_train, y_train)))
print('- Accuracy media en el conjunto test: {}'.format(clf.score(X_test, y_test)))
```
